```python
import numpy as np
import matplotlib.pyplot as plt
import os
%matplotlib inline
```

```python
data_type = 'bass'
folder = 'bass'
track_id = 'BryanAdams_SummerOf69'

base_path = os.path.join('../../multif0_ismir2017/training_data_with_blur', folder)
input_path = os.path.join(base_path, 'inputs', '{}_{}_input.npy'.format(track_id, data_type))
output_path = os.path.join(base_path, 'outputs', '{}_{}_output.npy'.format(track_id, data_type))
```

```python
X = np.load(input_path)
Y = np.load(output_path)
```

```python
plt.figure(figsize=(15, 15))
plt.subplot(2, 1, 1)
plt.imshow(X[1], origin='lower')
plt.axis('auto')

plt.subplot(2, 1, 2)
plt.imshow(Y, origin='lower')
plt.axis('auto')
```
